# Single-session inspection

Play a clip, scan static frames across the session, check pupil tracking, and view the pupil cloud in the eye-anchored frame. Helpers live in `eyevideo.py`.


```python
import eyevideo as ev
from IPython.display import Video

ev.ANIMAL_DIR = "/mnt/at-storageB1_I/EyeVideo/AT-B1NO1"   # <- animal folder
DATE = "2025-08-15"
sess = ev.session_for_date(DATE)
dur, nch = ev.session_duration(sess)
print(sess, f"\n{nch} chunks, {dur/60:.1f} min, frame size {ev.frame_size(sess)}")
```

## 1. Watch a clip
Random 20 s (real time). Re-run for a new window; `slowdown=8` for slow-mo, `start=` (s) for a fixed window.


```python
Video(ev.extract_clip(sess, duration=20.0), embed=True, width=640)
```

```python
Video(ev.extract_clip(sess, duration=5.0, slowdown=8), embed=True, width=640)   # slow motion
```

## 2. Static frames across the whole session
Drift check: 25 equally-spaced frames with a fixed reference grid + center crosshair.


```python
ev.drift_grid(sess, rows=5, cols=5, grid_step=100)
```

## 3. Pupil tracking on example frames
Robust ellipse (cyan outline + center) vs online centroid (orange ×).


```python
ev.show_tracking_examples(DATE, k=6)
```

## 4. Do the two trackers agree?
Left/right: ellipse-x vs online-x and ellipse-y vs online-y (dashed = identity). Then pupil x/y over the whole session (`n` frames).


```python
ev.tracker_agreement(DATE, n=200)
```

```python
ev.plot_pupil_xy(DATE, n=1000)     # n = number of equally-spaced frames
```

≥5 points around the eye opening, **right-click** to undo, then run `clk.save()`.
No-click fallback: `ev.show_mean_grid(DATE)` then `ev.set_landmarks(DATE, [(x,y), ...])`.
