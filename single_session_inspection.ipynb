# Single-session inspection

Look at one recording session: play a clip, scan static frames across the whole session,
check pupil tracking on example frames, and view the pupil-center cloud in the eye-anchored
coordinate frame. All helpers live in `eyevideo.py`.


```python
import eyevideo as ev
from IPython.display import Video

ev.ANIMAL_DIR = "/mnt/at-storageB1_I/EyeVideo/AT-B1NO1"   # <- animal folder

# choose a session by date (one session per day; longest if several)
DATE = "2025-08-15"
sess = ev.session_for_date(DATE)
dur, nch = ev.session_duration(sess)
print(sess, f"\n{nch} chunks, {dur/60:.1f} min, frame size {ev.frame_size(sess)}")
```

## 1. Watch a clip
Random 20 s (real time). Re-run for a new window; add `slowdown=8` for slow-mo or `start=` (s) for a fixed window.


```python
Video(ev.extract_clip(sess, duration=20.0), embed=True, width=640)
```

```python
Video(ev.extract_clip(sess, duration=5.0, slowdown=8), embed=True, width=640)   # slow motion
```

## 2. Static frames across the whole session
Drift check: 25 equally-spaced frames with a fixed reference grid (green) + center crosshair (red).


```python
ev.drift_grid(sess, rows=5, cols=5, grid_step=100)
```

## 3. Pupil tracking on example frames
Robust ellipse (cyan outline + center) vs the online centroid (orange ×).


```python
ev.show_tracking_examples(DATE, k=6)
```
